# Pre-processing Workflow for diat.barcode Database
diat.barcode Version: 12.4 [('2024-05-28-Diat.barcode_release-version 12.4.xlsx', v17.0)](https://entrepot.recherche.data.gouv.fr/file.xhtml?persistentId=doi:10.57745/O5BFF7&version=17.0), published on 05/06/2024.

This Jupyter Notebook contains all the pre-processing step to extract 2 separate dataset from the original diat.barcode database:
1. diat_barcode-sequence-v12.tsv: This file contains the sequence, sequence ID and the respective species nominal identification.
2. diar_barcode-taxonomy-v12.tsv: This file contains the taxonomic lineage of all the sequences and the respective sequence ID.

```python
import pandas as pd
# Imports
```

## Read and Exploratory Analysis

```python
diat = pd.read_csv('diat-barcode_v12.tsv', sep='\t')
```

```python
diat.head()
```

```python
# Get the dataframe shape
diat.shape
```

```python
# Remove columns and rows with all NaN values
diat.dropna(how='all', axis=1, inplace=True)
diat.dropna(how='all', axis=0, inplace=True)
```

```python
# No difference in shape -> No empty columns/rows were found
diat.shape
```

## Extract the sequence dataset

```python
# Select relevant columns
diat_barcode_seq = diat[['Sequence ID', 'Sequence', 'Species']]
```

```python
# Confirm the selection
diat_barcode_seq
```

```python
# Remove empty rows
diat_barcode_seq.dropna(how='all', axis=0, inplace=True)
```

```python
# Check shape
diat_barcode_seq.shape
```

```python
# Rename columns for machine readability
diat_barcode_seq.columns = ['seq_ID', 'sequence', 'species']
```

```python
# Write to file
diat_barcode_seq.to_csv('../../data/input_data/diat_barcode_sequence.csv', sep='\t', index=False)
```

## Extract Taxonomy Dataset

```python
col_names = diat.columns.to_list()
# col_names
```

### Extract Tradicional Taxonomy

```python
# Select columns refering to traditional taxonomy
diat_barcode_tax_1990 = diat[['Species',
                                'Genus',
                                'Family (following Round, Crawford & Mann 1990)',
                                'Order (following Round, Crawford & Mann 1990)',
                                'Class (following Round, Crawford & Mann 1990)',
                                'Phylum (following Algaebase 2018)',
                                'Subkingdom (following Algaebase 2018)',
                                'Kingdom (following Algaebase 2018)',
                                'Empire (following Algaebase 2018)']]
```

```python
# Rename columns for machine readability
diat_barcode_tax_1990.columns = ['Species',
                                'Genus',
                                'Family',
                                'Order',
                                'Class',
                                'Phylum',
                                'Subkingdom',
                                'Kingdom',
                                'Empire']
```

```python
diat_barcode_tax_1990.head()
```

```python
diat_barcode_tax_1990.shape
```

```python
diat_barcode_tax_1990.to_csv('../../data/input_data/diat_barcode_taxonomy_1990.tsv', sep='\t', index=False)
```

### Extract Modern Taxonomy

```python
# Select columns refering to modern taxonomy
diat_barcode_tax_2019 = diat[['Species',
                                'Genus',
                                'Family (following Adl et al. 2019)',
                                'Order (following Adl et al. 2019)',
                                'Subclass (following Adl et al. 2019)',
                                'Class (following Adl et al. 2019)',
                                'Subphylum (following Adl et al. 2019)',
                                'Subkingdom (following Adl et al. 2019)',
                                'Kingdom (following Adl et al. 2019)']]
```

```python
diat_barcode_tax_2019.columns = [['Species',
                                'Genus',
                                'Family',
                                'Order',
                                'Subclass',
                                'Class',
                                'Subphylum',
                                'Subkingdom',
                                'Kingdom']]
```

```python
diat_barcode_tax_2019.head()
```

```python
diat_barcode_tax_2019.shape
```

```python
diat_barcode_tax_2019.to_csv('../../data/input_data/diat_barcode_taxonomy_2019.tsv', sep='\t', index=False)
```
